```python
def dividi_blocchi(rows):
    blocks_texts= []
    current_block_text= []
    i=0
    for r in rows:
        if r.strip():
            current_block_text.append(r.strip())
        else:
            #print(f"{i}° blocco",current_block_text)
            if current_block_text:
                blocks_texts.append(current_block_text)
                current_block_text= []
            i+=1
    if current_block_text:
         blocks_texts.append(current_block_text)
    return blocks_texts
```
